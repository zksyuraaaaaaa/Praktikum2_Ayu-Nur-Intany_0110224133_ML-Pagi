```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. Load dataset
df = pd.read_csv("/content/gdrive/MyDrive/Praktikum2_MachineLearning/Data/day.csv")

# 2. Split data menjadi Training (80%) dan Testing (20%)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# 3. Dari Training, ambil 10% untuk Validation
train_df, val_df = train_test_split(train_df, test_size=0.1, random_state=42)

# 4. Cek jumlah data
print("Jumlah total data:", len(df))
print("Training:", len(train_df))
print("Validation:", len(val_df))
print("Testing:", len(test_df))

# 5. Tampilkan 5 data teratas
print("\nTraining Data:")
display(train_df.head())

print("\nValidation Data:")
display(val_df.head())

print("\nTesting Data:")
display(test_df.head())

# 6. Visualisasi pembagian data
import matplotlib.pyplot as plt

labels = ["Training", "Validation", "Testing"]
sizes = [len(train_df), len(val_df), len(test_df)]

plt.figure(figsize=(6,4))
plt.bar(labels, sizes)
plt.title("Distribusi Jumlah Data")
plt.ylabel("Jumlah Baris")
plt.show()


# Pie Chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Distribusi Proporsi Data (Pie Chart)")
plt.show()
```
